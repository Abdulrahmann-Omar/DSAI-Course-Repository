```python
import requests
import speech_recognition as sr
import pyttsx3

# Bing Speech API configuration
BING_SPEECH_API_KEY = "c3f79f94dd424961aef10161ab58ba50"
BING_SPEECH_API_ENDPOINT = "https://speech.platform.bing.com/speech/recognition/interactive/cognitiveservices/v1"

# Bing Text-to-Speech API configuration
BING_TTS_API_KEY = "YOUR_BING_TTS_API_KEY"
BING_TTS_API_ENDPOINT = "https://speech.platform.bing.com/synthesize"

# Speech recognition function
def recognize_speech(audio):
    headers = {
        "Content-Type": "audio/wav",
        "Authorization": "Bearer " + BING_SPEECH_API_KEY
    }
    response = requests.post(BING_SPEECH_API_ENDPOINT, headers=headers, data=audio)
    response.raise_for_status()
    return response.json()["DisplayText"]

# Text-to-speech function
def text_to_speech(text):
    headers = {
        "Content-Type": "application/ssml+xml",
        "Authorization": "Bearer " + BING_TTS_API_KEY
    }
    body = "<speak version='1.0' xmlns='http://www.w3.org/2001/10/synthesis' xml:lang='en-US'><voice xml:lang='en-US' xml:gender='Female' name='Microsoft Server Speech Text to Speech Voice (en-US, ZiraRUS)'>" + text + "</voice></speak>"
    response = requests.post(BING_TTS_API_ENDPOINT, headers=headers, data=body)
    response.raise_for_status()
    return response.content

# Initialize the speech recognizer
r = sr.Recognizer()

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Main loop
while True:
    try:
        # Listen for user input
        with sr.Microphone() as source:
            print("Listening...")
            audio = r.listen(source)

        # Recognize speech
        recognized_text = recognize_speech(audio.get_wav_data())

        # Print recognized text
        print("You said:", recognized_text)

        # Perform desired actions based on recognized text
        if recognized_text.lower() == "hello":
            response_text = "Hello! How can I assist you?"
        elif recognized_text.lower() == "goodbye":
            response_text = "Goodbye! Have a nice day!"
            break
        else:
            response_text = "I'm sorry, I didn't understand."

        # Convert the response text to speech
        response_audio = text_to_speech(response_text)

        # Play the response audio
        engine.setProperty("rate", 150)  # Adjust the speech rate (optional)
        engine.setProperty("volume", 1.0)  # Adjust the speech volume (optional)
        engine.save_to_file(response_audio, "response.wav")
        engine.runAndWait()

    except sr.UnknownValueError:
        print("Sorry, I could not understand audio.")
    except sr.RequestError as e:
        print("Could not request results from Bing Speech Recognition service; {0}".format(e))
    except requests.exceptions.HTTPError as e:
        print("HTTP Error occurred; {0}".format(e))
    except requests.exceptions.RequestException as e:
        print("Error occurred; {0}".format(e))
```
