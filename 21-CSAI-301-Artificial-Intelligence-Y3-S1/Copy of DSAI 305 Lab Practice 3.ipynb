## Data Cleaning

### Missing Values

```python

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python


# Dealing with missing values
# 1. LotFrontage: Numerical - Impute with median
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())

# 2. Alley, PoolQC, Fence, MiscFeature: Categorical - Impute with "None"
for col in ['Alley', 'PoolQC', 'Fence', 'MiscFeature']:
    df[col] = df[col].fillna('None')

# 3. MasVnrType, MasVnrArea: Categorical/Numerical - Impute with mode/median
df['MasVnrType'] = df['MasVnrType'].fillna(df['MasVnrType'].mode()[0])
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].median())

# 4. BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2: Categorical - Impute with "None"
for col in ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']:
    df[col] = df[col].fillna('None')

# 5. Electrical: Categorical - Impute with mode
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])

# 6. FireplaceQu: Categorical - Impute with "None"
df['FireplaceQu'] = df['FireplaceQu'].fillna('None')

# 7. GarageType, GarageYrBlt, GarageFinish, GarageQual, GarageCond: Categorical/Numerical - Impute with "None"/median
for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
    df[col] = df[col].fillna('None')
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())


# Verify missing values again
missing_values_after = df.isnull().sum()
print(missing_values_after[missing_values_after > 0])

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap After Imputation')
plt.show()
```

### Duplicates

```python

# Check for duplicates
duplicates = df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")

# Remove duplicates (if any)
df = df.drop_duplicates()

# Verify duplicates again
duplicates_after = df.duplicated()
print(f"Number of duplicate rows after removal: {duplicates_after.sum()}")
```

### Ouliers

```python

import numpy as np

def detect_outliers_zscore(data, threshold=3):
    z_scores = np.abs((data - data.mean()) / data.std())
    outliers = data[z_scores > threshold]
    return outliers

def detect_outliers_iqr(data, threshold=1.5):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    return outliers

numerical_cols = df.select_dtypes(include=np.number).columns
for col in numerical_cols:
    common_outliers = pd.Series(list(set(zscore_outliers.index) & set(iqr_outliers.index)))
    print(f"\nCommon outliers detected by both methods: \n{df.loc[common_outliers,col]}")
    print("-" * 50)
```

### Is there a need to convert data types?

```python

# Check data types of each column
print(df.dtypes)

# Identify columns that might need type conversion
# Example: converting 'MSSubClass' to object (categorical)
if df['MSSubClass'].dtype == 'int64':
    df['MSSubClass'] = df['MSSubClass'].astype(str)

# Example: converting 'OverallQual' and 'OverallCond' to categorical
for col in ['OverallQual', 'OverallCond']:
    if df[col].dtype in ['int64', 'float64']:
        df[col] = df[col].astype(str)

# Check data types again to verify changes
df.dtypes
```
