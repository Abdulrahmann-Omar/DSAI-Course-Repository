Simple barplot chart


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


t=pd.DataFrame({'color':['red','green','blue','yellow'],'total':[8,2,5,8]})

sns.barplot(data=t,x='color',y='total')

hue=["Red","Green","Blue","Yellow"]

sns.set_palette(hue)

sns.barplot(data=t,x='color',y='total')


plt.show()
```

```python

t=pd.DataFrame({'color':['red','green','blue','yellow'],'total':[8,2,5,8]})


hue=["Red","Green","Blue","Yellow"]

sns.set_palette(hue)

sns.barplot(data=t,x='total',y='color')


plt.show()
```
