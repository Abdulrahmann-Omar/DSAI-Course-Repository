```python
df = sns.load_dataset('exercise')
df
```

```python
df["kind"].value_counts()
```

```python
kinds = df["kind"].value_counts().index
values = df["kind"].value_counts().values

plt.pie(values, labels=kinds, autopct='%1.1f%%',
        startangle=140, colors=sns.color_palette('pastel'))
plt.axis('equal')
plt.title('Ratio of kinds')
plt.show()
```

```python
import seaborn as sns

# Sample data with datetime index
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
values = [10, 15, 8, 12, 9, 11, 14, 13, 16, 18]
data = pd.DataFrame({'Date': date_rng, 'Values': values})

# Create a line plot with datetime x-axis
sns.lineplot(data=data, x='Date', y='Values')

plt.title('Time Series Plot with Seaborn')
plt.xlabel('Date')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```

```python

data = sns.load_dataset('iris')
sns.scatterplot(data=data, x='sepal_length', y='sepal_width', hue='species')
plt.annotate('Iris Setosa', xy=(4.9, 3.1), xytext=(5.5, 4),
             arrowprops=dict(arrowstyle='->', lw=1.5),
             fontsize=12, color='red')

plt.title('Scatter Plot with Annotations')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python

data = {
    'Category': ['Revenue'],
    'Actual': [650],
    'Target': [750],
    'Minimum': [300],
    'Maximum': [1000]
}

import pandas as pd
df = pd.DataFrame(data)


fig, ax = plt.subplots(figsize=(6, 2))


sns.barplot(x='Actual', y='Category', data=df, color='lightblue', ax=ax)


plt.axvline(df['Target'][0], color='gray', linestyle='--')


plt.axvline(df['Minimum'][0], color='red', linestyle=':')
plt.axvline(df['Maximum'][0], color='green', linestyle=':')

plt.xlabel("Value")
plt.title("Bullet Graph Example")
```
