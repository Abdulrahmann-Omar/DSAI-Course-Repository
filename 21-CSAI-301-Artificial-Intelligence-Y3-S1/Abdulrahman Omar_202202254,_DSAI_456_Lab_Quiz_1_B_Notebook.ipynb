# Imports

```python
import librosa
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

# Reading & Plots (2)

```python
# Unzip Data
import numpy as np
```

```python
# Read Data
y1, sr1 = librosa.load('1.mp3')
y2, sr2 = librosa.load('2.mp3')
y3, sr3 = librosa.load('3.mp3')
y4, sr4 = librosa.load('4.mp3')
y5, sr5 = librosa.load('5.mp3')
y6, sr6 = librosa.load('6.mp3')
```

## Voice_1

```python

fig, ax = plt.subplots(nrows=1)
librosa.display.waveshow(y1, sr=sr1)
```

Interpreting: the loudness from s3 to s4 like shown

```python
f0_1, voiced_flag, voiced_probs = librosa.pyin(y1,
                                             sr=sr1,
                                             fmin=librosa.note_to_hz('C2'),
                                             fmax=librosa.note_to_hz('C7'))
times = librosa.times_like(f0_1, sr=sr1)
```

```python
import matplotlib.pyplot as plt
D = librosa.amplitude_to_db(np.abs(librosa.stft(y1)), ref=np.max)
fig, ax = plt.subplots()
img = librosa.display.specshow(D, x_axis='time', y_axis='log', ax=ax)
ax.set(title='pYIN fundamental frequency estimation')
fig.colorbar(img, ax=ax, format="%+2.f dB")
ax.plot(times, f0_1, label='f0', color='cyan', linewidth=3)
ax.legend(loc='upper right')
```

F0 for voice_1 = 256Hz

- **Use** **mel_spectrogram** function from librosa

```python
# Convert mel to db
S = np.abs(librosa.stft(y1))
librosa.power_to_db(S**2)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=2, sharex=True, sharey=True)
imgpow = librosa.display.specshow(S**2, sr=sr1, y_axis='log', x_axis='time',
                                  ax=ax[0])
ax[0].set(title='Power spectrogram')
ax[0].label_outer()
imgdb = librosa.display.specshow(librosa.power_to_db(S**2, ref=np.max),
                                 sr=sr1, y_axis='log', x_axis='time', ax=ax[1])
ax[1].set(title='Log-Power spectrogram')
fig.colorbar(imgpow, ax=ax[0])
fig.colorbar(imgdb, ax=ax[1], format="%+2.0f dB")
```

Interpreting: the light areas shows the sound energy is high, against dark areas

```python
D = np.abs(librosa.stft(y1))**2
S = librosa.feature.melspectrogram(S=D, sr=sr1)
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
S_dB = librosa.power_to_db(S, ref=np.max)
img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=sr1,
                         fmax=8000, ax=ax)
fig.colorbar(img, ax=ax, format='%+2.0f dB')
ax.set(title='Mel-frequency spectrogram')
```
