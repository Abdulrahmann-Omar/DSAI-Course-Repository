```python
import matplotlib.pyplot as plt
```

```python
file1 = open('model_1.txt', 'r')

Lines = file1.readlines()

X = []
Y = []
Z = []

for i in range (0,2770):
  
  SS = Lines[i].split(' ')
  X.append(float(SS[0]))
  Y.append(float(SS[1]))
  Z.append(float(SS[2]))
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter(X,Y,Z, s = 2.7)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```

```python
xMin = -0.4
xMax = 0.4
yMin = 0.11
yMax = 0.19
zMin = -0.4
zMax = 0.4

XN = []
YN = []
ZN = []
XL = []
YL = []
ZL = []

for i in range (len(X)):
  if ((X[i] >= xMin and X[i] <= xMax)
  and (Y[i] >= yMin and Y[i] <= yMax)
  and (Z[i] >= zMin and Z[i] <= zMax)):
    XL.append(X[i])
    YL.append(Y[i])
    ZL.append(Z[i])
  else:
    XN.append(X[i])
    YN.append(Y[i])
    ZN.append(Z[i])
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter(XN,YN, ZN, color='red' , s = 0.7) #out
ax.scatter(XL,YL, ZL, color='blue', s = 2.7) #in
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.view_init(30, 10)

plt.draw()
```
