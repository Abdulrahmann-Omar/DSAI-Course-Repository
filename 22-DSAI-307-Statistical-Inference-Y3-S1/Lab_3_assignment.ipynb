### K -NEAREST NEIGHBOR (K-NN)CLASSIFIER

# How it works?
[1] Store training data: The algorithm stores all the training data points in memory.

[2] Calculate distances: When a new data point (query point) needs to be classified or
predicted, the algorithm calculates its distance to all training points.

[3] Find nearest neighbors: The K nearest neighbors to the query point are identified based
on the calculated distances.

[4] Make prediction:

Classification: The majority class among the K nearest neighbors is assigned to the
query point.

Regression: The average value of the target variable among the K nearest neighbors
is assigned to the query point.
