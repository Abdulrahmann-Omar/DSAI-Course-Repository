```r
lambda <- 1 / 2
n_samples <- 1500
```

```r
generate_sample_means <- function(n) {
  sample_means <- replicate(n_samples, mean(rexp(n, rate = lambda)))

  hist(sample_means,
       breaks = 30,
       probability = TRUE,
       main = paste(" Sample Means n =", n),
       xlab = "Sample Mean",
       col = "skyblue",
       border = "white")

  lines(density(sample_means), col = "red", lwd = 2)

  mean_theoretical <- 2  #  mean of exp dist
  sd_theoretical <- sqrt(2^2 / n)  # S d of sample mean
  curve(dnorm(x, mean = mean_theoretical, sd = sd_theoretical),
        col = "darkblue", lwd = 2, add = TRUE, lty = 2)
}
```

```r
par(mfrow = c(3, 1))
generate_sample_means(10)
generate_sample_means(50)
generate_sample_means(100)
```

For n=10:
sample size is small. that make the dist not look very normal


For n=50: the dist tends to be normal distribution more closely.


For
n=100: the dist tends to be normal distribution more and more closely.
