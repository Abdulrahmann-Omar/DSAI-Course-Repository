# Problem 1: Calculating the Statistical Measures in R


```r
summary(trees)
```

```r
# print number of columns
print (ncol(trees))

# print number of rows
print(nrow(trees))
```

```r
install.packages("modeest")
```

```r
# Import the library
library(modeest)

# Compute the mode for each column in the dataset
modes = sapply(trees, mfv)

# Print the modes
print(modes)
```

```r
mean = sapply(trees, mean)

print(mean)
```

```r
var = sapply(trees, var)

print(var)
```

```r
sd = sapply(trees, sd)

print(sd)
```

```r
median = sapply(trees, median)

print(median)
```

# Problem 2: Visualizing Data Distributions

```r
library(ggplot2)
```

```r
ggplot(data = trees) +
labs(title = "trees Data Plot")
```

```r
# Histogram plot
ggplot(data = trees, aes(x = Height)) +
geom_histogram(binwidth = 5) +
labs(title = "Histogram of Horsepower",
	x = "Horsepower",
	y = "Count")
  #approximatly follows the normal distribution
```

```r
# Boxplot for Height
ggplot(trees, aes(y = Height)) +
  geom_boxplot(fill = "blue") +
  labs(title = "Boxplot of Height", y = "Height (ft)")
  # No outliers
```

```r
# Histogram plot
ggplot(data = trees, aes(x = Volume )) +
geom_histogram(binwidth = 5) +
labs(title = "Histogram of Horsepower",
	x = "Horsepower",
	y = "Count")
  #highly right-skewed
```

```r
ggplot(trees, aes(y = Volume)) +
  geom_boxplot(fill = "yellow") +
  labs(title = "Boxplot of Volume", y = "Volume (cubic ft)")
  #most trees have a volume between 10 and 50  feet
```

```r
# Histogram plot
ggplot(data = trees, aes(x =Girth )) +
geom_histogram(binwidth = 5) +
labs(title = "Histogram of Horsepower",
	x = "Horsepower",
	y = "Count")
  #Girth appears slightly right-skewed which means most trees have a moderate girth
```

```r
ggplot(trees, aes(y = Girth)) +
  geom_boxplot(fill = "lightgreen") +
  labs(title = "Boxplot of Girth", y = "Girth (inches)")
  # few outliers
```
