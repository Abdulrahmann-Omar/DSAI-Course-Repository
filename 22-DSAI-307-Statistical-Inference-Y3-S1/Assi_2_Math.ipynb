```python
#Q2
from scipy.optimize import linprog

# Objective function
c = [-1, 8]  # We need to  find max for c

# inequality constraints
A = [
    [-3, -2],
    [1, -1],
    [9, 7],
    [3, 7],
    [-2, 5],
]

b = [-6, 6, 108, 70, 35]

# Bounds for x and y
x_bounds = (0, None)
y_bounds = (0, None)

# Solve for feasible region vertices by maximizing and minimizing x and y
# Run linprog in both directions to explore boundaries and feasible points

# Minimize x to find lower boundary
res_min_x = linprog([1, 0], A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')
# Maximize x (equivalently minimize -x)
res_max_x = linprog([-1, 0], A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Minimize y
res_min_y = linprog([0, 1], A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')
# Maximize y (equivalently minimize -y)
res_max_y = linprog([0, -1], A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Collect feasible points from boundary solutions
feasible_points = [
    (res_min_x.x[0], res_min_x.x[1]),
    (res_max_x.x[0], res_max_x.x[1]),
    (res_min_y.x[0], res_min_y.x[1]),
    (res_max_y.x[0], res_max_y.x[1])
]

# Remove any duplicates (rounding to handle precision issues)
feasible_points = list(set((round(x, 5), round(y, 5)) for x, y in feasible_points))

feasible_points
```

```python
#Q5
from scipy.optimize import linprog

# objective function
c = [-3, -2]

# inequality constraints
A = [
    [4, -1],
    [4, 3],
    [4, 1]
]

# Right-hand side values
b = [8, 12, 8]

# Bounds
x_bounds = (0, None)
bounds = [x_bounds, x_bounds]

# Solve
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')


print("Optimal solution found:")
print("x1 =", result.x[0])
print("x2 =", result.x[1])
print("Maximum value of objective function =", -result.fun)  # Re-negate to get the maximized objective value

```
