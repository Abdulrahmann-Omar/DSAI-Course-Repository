#** **Data Validation:**

```python
!pip install pandera
import pandera as pa
import pandas as pd
from pandera import Column, DataFrameSchema, Check, Index
```

```python

# Define the validation schema
schema = DataFrameSchema({
    'MOVIES': Column(str, Check.str_length(min_value=1), nullable=False),
    'GENRE': Column(str, Check.str_length(min_value=1), nullable=False),
    'RATING': Column(float, Check.in_range(0, 10), nullable=True),
    'ONE-LINE': Column(str, Check.str_length(min_value=1), nullable=True),
    'STARS': Column(str, Check.str_length(min_value=1), nullable=True),
})


validated_df = schema.validate(df)
print("Data validation successful.")
# Proceed with further data processing or save the validated dataframe
```

```python
validated_df
```

```python

# encryption key
key = Fernet.generate_key()
cipher_suite = Fernet(key)

# Encrypt sensitive information
sensitive_columns = ['RATING','STARS','VOTES']
for column in sensitive_columns:

        df[column] = df[column].apply(lambda x: cipher_suite.encrypt(str(x).encode()))

print("completed.")
```
