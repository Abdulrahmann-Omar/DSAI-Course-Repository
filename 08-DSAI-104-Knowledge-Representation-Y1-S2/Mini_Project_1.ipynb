```python
import itertools

def check_conditions(assignment):
    A, B, C, D, E, F = assignment

    cond_A = (not A or not B) and (A or B)

    cond_B = (not B or (C and D)) and (B or not (C and D))

    cond_C = (not C or (not A or not E)) and (C or (A and E))

    cond_D = (not D or ((not A or B) and (not B or A))) and (D or ((A and not B) or (not A and B)))

    cond_E = (not E or (not A or not B or not C or not D or not F)) and (E or (A and B and C and D and F))

    cond_F = (not F or (D and not E)) and (F or (not D or E))

    total_truth = sum(assignment)

    return all([cond_A, cond_B, cond_C, cond_D, cond_E, cond_F])

possible_assignments = itertools.product([True, False], repeat=6)
solutions = []

for assignment in possible_assignments:
    if check_conditions(assignment):
        solutions.append(assignment)



print(f"A: {solutions[0][0]}")
print(f"B: {solutions[0][1]}")
print(f"C: {solutions[0][2]}")
print(f"D: {solutions[0][3]}")
print(f"E: {solutions[0][4]}")
print(f"F: {solutions[0][5]}")
```
