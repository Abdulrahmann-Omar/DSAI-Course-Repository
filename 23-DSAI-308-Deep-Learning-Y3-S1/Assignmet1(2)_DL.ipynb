```python
#First Unzip the Dataset File
import zipfile
with zipfile.ZipFile('/content/archive.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
import os
import shutil
import numpy as np
```

```python
# Define paths
rootdir = '/content/'
train_path = os.path.join(rootdir, 'Training')
test_path = os.path.join(rootdir, 'Testing')
classes_dir = ['yes', 'no']  # Assuming these are the class names

# Create directories for train and test
for i in classes_dir:
    if not os.path.exists(os.path.join(rootdir, 'train', i)):
        os.makedirs(os.path.join(rootdir, 'train', i))
    if not os.path.exists(os.path.join(rootdir, 'test', i)):
        os.makedirs(os.path.join(rootdir, 'test', i))

# Copy files from Training to train
for class_name in classes_dir:
    train_source = os.path.join(train_path, class_name)
    test_source = os.path.join(test_path, class_name)

    # Check if the source directories exist
    if not os.path.exists(train_source):
        print(f"Directory {train_source} does not exist. Check your dataset structure.")
        continue
    if not os.path.exists(test_source):
        print(f"Directory {test_source} does not exist. Check your dataset structure.")
        continue

    # Copy training files
    train_files = os.listdir(train_source)
    for file_name in train_files:
        src = os.path.join(train_source, file_name)
        dst = os.path.join(rootdir, 'train', class_name, file_name)
        shutil.copy(src, dst)

    # Copy testing files
    test_files = os.listdir(test_source)
    for file_name in test_files:
        src = os.path.join(test_source, file_name)
        dst = os.path.join(rootdir, 'test', class_name, file_name)
        shutil.copy(src, dst)

print("Dataset organized successfully!")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
import os

rootdir = '/content/'
classes_dir = ['yes', 'no']

for class_name in classes_dir:
    im_src = os.path.join(rootdir, 'train', class_name)
    FileNames = os.listdir(im_src)
    print(f"Number of files in {im_src}: {len(FileNames)}")
```

```python
fig, ax = plt.subplots(2,5, figsize=(30,15))
fig.suptitle('Brain MRI  Categories',color='magenta',fontsize=40)
for i in range(2):
    for j in range(5):
        im_src = rootdir +'train/' + classes_dir[i] + '/'
        FileNames = os.listdir(im_src)
        np.random.shuffle(FileNames)
        im_src = str(im_src + FileNames[j])
        imag = img.imread(im_src)
        ax[i, j].imshow(imag)
        ax[i, j].set_title(classes_dir[i], fontsize=20)
```
